```python
# Preprocessing
train = ImageDataGenerator(rescale=1/255)
image_generator = ImageDataGenerator(rescale=1/255, validation_split=0.2)    

train_dataset = image_generator.flow_from_directory(batch_size=32,
                                                 directory="../input/huy-datasets-fruits10/huy_datasets_fruits10/train",
                                                 shuffle=True,
                                                 target_size=(150, 150), 
                                                 subset="training",
                                                 class_mode='categorical')

validation_dataset = image_generator.flow_from_directory(batch_size=32,
                                                 directory="../input/huy-datasets-fruits10/huy_datasets_fruits10/validation",
                                                 shuffle=True,
                                                 target_size=(150, 150), 
                                                 subset="validation",
                                                 class_mode='categorical')
```

```python
# Reshape Data
import glob
apple          = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/apple/*.*')
banana         = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/banana/*.*')
beetroot       = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/beetroot/*.*')
bellpepper     = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/bell pepper/*.*')
cabbage        = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/cabbage/*.*')
cappsicum      = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/capsicum/*.*')
carrot         = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/carrot/*.*')
cauliflower    = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/cauliflower/*.*')
chilipepper    = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/chilli pepper/*.*')
corn           = glob.glob('../input/huy-datasets-fruits10/huy_datasets_fruits10/train/corn/*.*')

data = []
labels = []

for i in apple:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(0)
for i in banana:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(1)
for i in beetroot:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(2)
for i in bellpepper:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(3)

for i in cabbage:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(4)
for i in cappsicum:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(5)
for i in carrot:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(6)
for i in cauliflower:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(7)
for i in chilipepper:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(8)
for i in corn:   
    image=tf.keras.preprocessing.image.load_img(i, color_mode='rgb', 
    target_size= (150,150))
    image=np.array(image)
    data.append(image)
    labels.append(9)
```

```python
data = np.array(data)
labels = np.array(labels)
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2,random_state=50)
```

```python
# Kich thuoc cac tap du lieu 
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train = x_train.reshape((x_train.shape[0],150,150,3))
x_train = x_train.astype('float32')/255

x_test = x_test.reshape((x_test.shape[0],150,150,3))
x_test = x_test.astype('float32')/255

y_train = to_categorical(y_train,10)
y_test = to_categorical(y_test,10)
```

```python
# Khoi tao model 
from keras.layers import Conv2D, MaxPooling2D
model = Sequential()
model.add(Conv2D(32,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same',input_shape=(150,150,3)))
model.add(Conv2D(32,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same')) # 64 lan tich chap
model.add(Conv2D(64,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(128,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same')) # 128 lan tich chap
model.add(Conv2D(128,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(256,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same')) # 256 lan tich chap
model.add(Conv2D(256,(3,3), activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D(2,2))

from keras.layers import Dense, Activation, Flatten
model.add(Flatten())
model.add(Dense(128, activation = 'relu', kernel_initializer='he_uniform'))
model.add(Dense(10))
model.summary()
```

```python
model.compile(loss='mse',optimizer=RMSprop(),metrics=['accuracy'])
# Training model
history=model.fit(train_dataset,batch_size=164,epochs=50,validation_data=validation_dataset)
```

```python
# Save model
from tensorflow.keras.models import load_model
model.save('huyFruit10.h5')
model_fruit10 = load_model('huyFruit10.h5')
```

```python
# Results
score = model.evaluate(x_test,y_test, verbose=0)
print("Loss = ", score[0])
print("accuracy = ", score[1])
```

```python
# Vẽ lại quá trình học
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['train','Validation'])
plt.show()
```
